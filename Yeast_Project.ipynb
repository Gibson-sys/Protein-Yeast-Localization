```python
X = dataset.iloc[:,:-1]
y = dataset["name"]
from imblearn.over_sampling import SMOTE

smote = SMOTE(k_neighbors=2)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
### SPLIT THE DATASET
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size = 0.2,random_state = 42)
x_train
```

```python
new_x_train = x_train[:2000]
new_y_train = y_train[:2000]
# validation set
x_val = x_train[2000:]
y_val = y_train[2000:]
```

```python
#CREATING A NEAURAL NETWORK WORK MODEL 
import tensorflow as tf

# set seed
tf.random.set_seed(42)

#FIRST MODEL
model = tf.keras.Sequential([
    tf.keras.layers.Dense(400, input_shape=(8,), activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(400, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(300, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(128, activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(10, activation="softmax")
])

model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

histo = model.fit(new_x_train,new_y_train,validation_data = (x_val,y_val),epochs = 500)
```

```python

pd.DataFrame(histo.history).plot(figsize=(15,5))
plt.show();
```

```python
model.evaluate(x_test,y_test)
```

```python
y_preed = model.predict(x_test)
new_pred = np.argmax(y_preed,axis = 1)
print(new_pred[:100])
print("=====================================")
print("=====================================")
print(list(y_test[:100]))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,precision_score,f1_score,recall_score,accuracy_score
```

```python
print(classification_report(y_test,new_pred))
print("👍👍👍👍👍👍👍👍👍Confusion Matrix👍👍👍👍👍👍👍👍👍👍👍")
print(confusion_matrix(y_test,new_pred))
```

```python

from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Create the confusion matrix
cm = confusion_matrix(y_test, new_pred)

# Initialize a figure
fig, ax = plt.subplots(figsize=(10, 7))

# Function to update the heatmap
def update(frame):
    ax.clear()
    temp_cm = cm.copy()
    temp_cm = temp_cm + np.random.randint(-1, 2, size=temp_cm.shape)  # Adding small random values
    sns.heatmap(temp_cm, annot=True, fmt="d", cbar=False, ax=ax)
    ax.set_title(f"Confusion Matrix Animation (Frame {frame})")
    ax.set_xlabel("Predicted Labels")
    ax.set_ylabel("True Labels")

# Create an animation
anim = FuncAnimation(fig, update, frames=20, interval=500)

# Display the animation in a Jupyter Notebook
HTML(anim.to_jshtml())  # Display in Jupyter Notebook

# To save the animation as a file:
anim.save('matrix_animation.gif', writer='pillow')
plt.show()
```
